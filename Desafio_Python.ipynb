```python
import ipywidgets as widgets
from IPython.display import display, clear_output
from ipywidgets import interact, Dropdown

# Mover o dicion√°rio para o escopo global
descricoes = {
    "Relatorio de juros em planilha Excel": "üìä Calculo de juros sobre saldos usando colunas da planilha",
    "Tabela de clientes com limite de credito": "üë• Lista dados de clientes e campo com limite de credito",
    "Consulta SQL de saldo por cliente": "üíæ Filtra tabela de contas para mostrar saldo de cada cliente",
    "Consulta SQL de historico de transacoes": "üìù Usa SQL para buscar transacoes antigas de uma conta",
    "Planilha Excel de simulacao de parcelas de emprestimo": "üí∞ Calcula valor das parcelas com juros sobre o emprestimo",
    "Relatorio em Excel com resumo de gastos do cliente": "üìà Mostra resumo mensal dos gastos do cliente no banco",
    "Consulta SQL com total de compras no cartao": "üí≥ Soma compras registradas no cartao em uma tabela",
    "Consulta SQL listando emprestimos em aberto": "‚ö†Ô∏è Retorna emprestimos ativos para analise de risco"
}

def descricao_uso(tecnica):
    # A fun√ß√£o agora pode acessar o dicion√°rio global
    return descricoes.get(tecnica, "‚ùå Tecnica desconhecida")

# Criar interface
print("=" * 50)
print("üè¶ BANKTECH - SISTEMA DE T√âCNICAS BANC√ÅRIAS")
print("=" * 50)

tecnica_select = Dropdown(
    options=list(descricoes.keys()),
    description='T√©cnica:',
    style={'description_width': 'initial'}
)

output = widgets.Output()

def on_change(change):
    with output:
        clear_output()
        print("\n" + "=" * 40)
        print("üìã RESULTADO:")
        print("=" * 40)
        print(f"T√©cnica: {tecnica_select.value}")
        print(f"Descri√ß√£o: {descricao_uso(tecnica_select.value)}")
        print("=" * 40)

tecnica_select.observe(on_change, names='value')

display(tecnica_select)
display(output)

# Mostrar resultado inicial
with output:
    print("\n" + "=" * 40)
    print("üìã RESULTADO:")
    print("=" * 40)
    print(f"T√©cnica: {tecnica_select.value}")
    print(f"Descri√ß√£o: {descricao_uso(tecnica_select.value)}")
    print("=" * 40)
```
